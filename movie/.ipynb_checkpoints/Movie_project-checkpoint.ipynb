# Project 1: Explanatory Data Analysis & Data Presentation (Movies Dataset)

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
from IPython.display import HTML
from wordcloud import WordCloud
df = pd.read_csv("movies_complete.csv",index_col="id")
df.head()
```

__Some additional information on Features/Columns__:
* **id:** ID của movie (clear/unique identifier).
* **title:** Tiêu đề
* **tagline:**  tagline của movie.
* **release_date:** Thời gian chiếu.
* **genres:** Thể loại phim.
* **belongs_to_collection:** Thông tin của movie series/nhượng quyền thương mại.
* **original_language:** Ngôn ngữ gốc
* **budget_musd:** Toàn bộ ngân sách.
* **revenue_musd:** Toàn bộ doanh thu.
* **production_companies:** Những công ty hợp tác.
* **production_countries:** Các quốc gia nơi phim được quay.
* **vote_count:** Số lượng vote (TMDB).
* **vote_average:** Trung bình rateing.
* **popularity:** Điểm nổi tiếng (TMDB).
* **runtime:** Thời lượng phim.
* **overview:** Tổng quan phim.
* **spoken_languages:** Ngôn ngữ trong phim.
* **poster_path:** liên kết URL poster của phim.
* **cast:** Diễn viên trong phim.
* **cast_size:** Số lượng diễn viên.
* **director:** Đạo diễn.
* **crew_size:** Quy mô đoàn làm phim (bao gồm đạo diễn, không bao gồm diễn viên).

## Preprocessing

```python
df.nunique()
```

```python
col_convert = ["original_language","genres","belongs_to_collection",
               "production_countries",
               "runtime","spoken_languages","cast_size","crew_size"]
df[col_convert] = df[col_convert].astype("category")
```

```python
df.info()
```

## 1. The best  movies.


- Profit (=Revenue - Budget)
- Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 

__Movies Top 5 - Highest Revenue__

```python
def HTML_image(df):
    return HTML(df.to_html(escape = False))

highest_revenue = df[["title","poster_path","revenue_musd"]].sort_values("revenue_musd",ascending = False).head()
HTML_image(highest_revenue)
```

__Movies Top 5 - Highest Budget__

```python
highest_budget = df[["title","poster_path","budget_musd"]].sort_values("budget_musd",ascending = False).head()
HTML_image(highest_budget)
```

__Movies Top 5 - Highest Profit__

```python
df["profit"]  = df.revenue_musd.sub(df.budget_musd)
highest_profit = df[["title",'poster_path','profit']].sort_values('profit',ascending = False).head()
HTML_image(highest_profit)
```

__Movies Top 5 - Highest ROI__

```python
df['ROI'] = df.revenue_musd.div(df.budget_musd)
highest_roi = df[df.budget_musd>10][["title",'poster_path','ROI']].sort_values("ROI",ascending = False).head()
HTML_image(highest_roi)
```

__Movies Top 5 - Lowest ROI__

```python
lowest_roi = df[df.budget_musd>10][["title",'poster_path','ROI','budget_musd']].sort_values("ROI").head()
HTML_image(lowest_roi)
```

__Movies Top 5 - Most Votes__

```python
highest_vote = df[["title",'poster_path','vote_count']].sort_values('vote_count',ascending = False).head()
HTML_image(highest_vote)
```

__Movies Top 5 - Highest Rating__

```python
highest_rating = df[["title",'poster_path','vote_average']].sort_values(['vote_average','poster_path'],ascending = False).head()
HTML_image(highest_rating)
```

__Movies Top 5 - Most Popular__

```python
most_popular = df[["title",'poster_path','popularity']].sort_values('popularity',ascending = False).head()
HTML_image(most_popular)
```

__Movies Top 5 - Tagline - Overview__

```python
def wordcloud_trans(series):
    overview_prep = series.dropna()
    overview_wordcloud = ''.join(overview_prep)
    overview_wordcloud = WordCloud(width=2000,height=1000,margin = 1,max_words=500,background_color='salmon').generate(overview_wordcloud)
    plt.figure(figsize=(15,10))
    plt.axis('off')
    plt.imshow(overview_wordcloud)
```

```python
wordcloud_trans(df.overview)
```

```python
wordcloud_trans(df.tagline)
```

## 2. Find your next Movie

__Search 1: Science Fiction Action Movie with Bruce Willis (sorted from high to low Rating)__

```python
mask_genres_1 = df.genres.str.contains("Science",na = False) & df.genres.str.contains("Action",na = False) & df.genres.str.contains("Fiction",na = False)
mask_actor_1 = df['cast'].str.contains("Bruce Willis",na = False)
#mask = mask_genres & mask_actor
search_1 = df.loc[mask_actor_1 & mask_genres_1,["title",'poster_path','vote_average']].sort_values('vote_average',ascending = False)
HTML_image(search_1)
```

__Search 2: Movies with Uma Thurman and directed by Quentin Tarantino (sorted from short to long runtime)__

```python
df_search_2 = df[df.director == 'Quentin Tarantino']
mask_actor_2 = df_search_2.cast.str.contains('Uma Thurman', na = False)
search_2 = df_search_2.loc[mask_actor_2 ,["title",'poster_path','runtime']].sort_values('runtime')
HTML_image(search_2)
```

__Search 3: Most Successful Pixar Studio Movies between 2010 and 2015 (sorted from high to low Revenue)__

```python
mask_company_3 = df.production_companies.str.contains('Pixar',na = False)
mask_date_3 = df.release_date.between('2010-01-01','31-12-2015')
search_3 = df.loc[mask_company_3 & mask_date_3 ,["title",'poster_path','revenue_musd']].sort_values('revenue_musd',ascending = False)
HTML_image(search_3)
```

## Are Franchises more successful?

__Franchise vs. Stand-alone: Average Revenue__

```python
franchise = df.belongs_to_collection.notna()
stan_alone = df.belongs_to_collection.isna()
franchise_average_revenue = df[franchise].revenue_musd.mean()
stan_alone_average_revenue = df[stan_alone].revenue_musd.mean()
plt.bar(x = ['Franchise','Stand-alone'],height = [franchise_average_revenue,stan_alone_average_revenue],color = ['Salmon','lightblue'])
plt.ylabel('Million Dollar')
plt.title('Average Revenue');
```

__Franchise vs. Stand-alone: Return on Investment / Profitability (median)__

```python
franchise_ROI = df[franchise].ROI.median()
stan_alone_ROI = df[stan_alone].ROI.median()
plt.bar(x = ['Franchise','Stand-alone'],height = [franchise_ROI,stan_alone_ROI],color = ['Salmon','lightblue'])
plt.ylabel('Million Dollar')
plt.title('Return on Investment / Profitability (median)');
```

__Franchise vs. Stand-alone: Average Budget__

```python
franchise_budget = df[franchise].budget_musd.mean()
stan_alone_budget = df[stan_alone].budget_musd.mean()
plt.bar(x = ['Franchise','Stand-alone'],height = [franchise_budget,stan_alone_budget],color = ['Salmon','lightblue'])
plt.ylabel('Million Dollar')
plt.title('Average Budget');
```

__Over view__


```python
summary = df.groupby(franchise).agg({"budget_musd":"mean","revenue_musd":"mean",
                 "vote_count":"mean","vote_average":"mean",
                 "ROI":"median","profit":"mean"})
x = np.arange(len(summary.columns))
plt.figure(figsize=(15,10))
a = plt.bar(x = x, height=summary.values[1],width = 0.35,label= "Franchise",color = "salmon")
b = plt.bar(x = x+0.35,height=summary.values[0],width = 0.35,label = "Stand-alone",color = 'lightblue')
for i, value in enumerate(summary.values[1]):
    plt.text(i, value + 0.05, f'{value:.2f}', ha='center', va='bottom')
    
for i, value in enumerate(summary.values[0]):
    plt.text(i+0.35, value + 0.05, f'{value:.2f}', ha='center', va='bottom')
plt.legend()
plt.title("Summary",fontsize = 35)
plt.xticks(ticks = x+0.35/2, labels = summary.columns);
```

## Most Successful Actors

```python
cast = df.cast.str.split('|',expand = True)
cast_df = pd.DataFrame(cast.stack().reset_index(level = 1,drop = True))
cast_df.columns = ['actor']
cast_df = pd.merge(df[['title','genres','budget_musd','revenue_musd','vote_count','vote_average']],cast_df,how = 'left',
        left_on='id',right_on='id')
cast_df.head()
```

- __total casting of movies__

```python
top_10_casting = cast_df.groupby('actor').count().nlargest(10,'title')
plt.figure(figsize=(15,10))
plt.bar(x = top_10_casting.index,height = top_10_casting.title)
for i,v in enumerate(top_10_casting.title):
     plt.text(x = i,y = v, s = str(v), ha='center', va='bottom')
plt.title("the number of movies in which the actor participated")
plt.xticks(fontsize = 9);
```

- __total revenue__

```python
top_10_casting = cast_df.groupby('actor').sum().nlargest(10,'revenue_musd')
plt.figure(figsize=(15,10))
plt.bar(x = top_10_casting.index,height = top_10_casting.revenue_musd)
for i,v in enumerate(top_10_casting.revenue_musd):
     plt.text(x = i,y = v, s = f"{v:.2f}" , ha='center', va='bottom')
plt.title("Total revenue")
plt.xticks(fontsize = 9);
```
