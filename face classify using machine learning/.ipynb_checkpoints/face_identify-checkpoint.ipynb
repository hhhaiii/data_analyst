```python
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
from sklearn.datasets import fetch_olivetti_faces
from sklearn.model_selection import cross_val_score,GridSearchCV
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
x = fetch_olivetti_faces().data
y = fetch_olivetti_faces().target
len(x),len(np.unique(y))
```

* X gồm 400 ảnh 64x64 pixal
* Y có 40 giá trị độc lập từ 0-39 tương đương với 40 khuôn mặt khác nhau
* __=> 10 ảnh đặc trưng cho 1 khuôn mặt__


```python
fig,ax = plt.subplots(nrows = 5,ncols = 8,figsize = (14,8))
ax=ax.flatten()
for i in range(40):
    ax[i].imshow(x[10*i+9].reshape(64,64))
    ax[i].set_xticks([])
    ax[i].set_yticks([])
    ax[i].set_title("Face %d" %(i+1))
```

```python
plt.plot(PCA().fit(x).explained_variance_)
plt.xlabel("Feature")
plt.ylabel("Variances (%)");
```

### Từ sơ đồ trên có thể thấy được 100 feature đầu tiên ảnh hưởng nhiều nhất đến tập dữ liệu gốc. Những feature còn lại ảnh hưởng rất ít 

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
pca = PCA(n_components=100,whiten=True)
pca.fit(x_train)
pca.transform(x_train)
pca.transform(x_test)
x_pca = pca.fit_transform(x)
models = (("GaussianNB",GaussianNB()),("SVC",SVC()),("LogisticRegression",LogisticRegression()))
for name,model in (models):
    accuracy = cross_val_score(model,x_pca,y,cv = 5)
    print("%s:%f"%(name, accuracy.mean()))
```

### LogisticRegression(), điều chỉnh tham số  GridSearchCV()

```python
grid_log_reg ={
    'C': [10, 1, 0.1],
    'solver': ['lbfgs', 'liblinear'],
    'max_iter': [1000],
}

model_log_reg = LogisticRegression()
gs_log_reg = GridSearchCV(estimator=model_log_reg,param_grid=grid_log_reg,cv=5,verbose=2)
gs_log_reg.fit(x_train,y_train) 
```

```python
gs_log_reg.best_params_
```

```python
gs_log_reg.score(x_test,y_test)
```

```python
y_pre = gs_log_reg.predict(x_test)
```

```python
print("Kết quả dự đoán %d / %d" %(np.sum(y_test==y_pre),len(y_test)))
```
